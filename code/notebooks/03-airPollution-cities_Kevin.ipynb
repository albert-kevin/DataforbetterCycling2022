```python
# pollution ratio
city_pollute=airpo.groupby(['City'],as_index=False).Value.sum()
city_pollute['long']=long_lat.long
city_pollute['lat']=long_lat.lat
city_pollute['total']=sum(city_pollute.Value)
city_pollute['pollute_ratio']=city_pollute.Value/city_pollute['total']
city_pollute['Location']=airpo['Location']
city_pollute['year']=airpo['Year']
city_pollute.head()
```

```python
# pollution ratio 2020
airpo2020=airpo[airpo.Year==2020]
city_p2020=airpo2020.groupby(['City'],as_index=False).Value.sum()
city_p2020['long']=long_lat.long
city_p2020['lat']=long_lat.lat
city_p2020['total']=sum(city_p2020.Value)
city_p2020['pollute_ratio']=city_p2020.Value/city_p2020['total']
city_p2020['Location']=airpo['Location']
```

```python
#heat map 
def generateBaseMap(loc, zoom=9, tiles='OpenStreetMap', crs='ESPG2263'):
    return folium.Map(location=loc,
                   control_scale=True, 
                   zoom_start=zoom,
                   tiles=tiles)

MAP = [city_pollute.lat[0],city_pollute.long[0]] 
  
base_map = generateBaseMap(MAP)

city_pollute['Values']=city_pollute['Value']/100
map_values1 = city_pollute[['lat','long','pollute_ratio']]

data = map_values1.values.tolist()
           
hm = HeatMap(data,gradient={0.009: 'blue', 0.02: 'lime', 0.05: 'yellow', 0.08: 'orange', 0.1: 'red'}, 
                min_opacity=0.05, 
                max_opacity=0.9, 
                radius=25,
                use_local_extrema=False)#.add_to(base_map)

#base_map
base_map.add_child(hm)
```

```python
# interactive map
def plot_map(df_select, i_df, i_scale):
    zoom_factor = 8 # inital map size
    radius_scaling = i_scale # scaling of bubbles

    my_map = folium.Map(location=[50.5,4], zoom_start=zoom_factor)

    for i in range(0,df_select.shape[0]):
        folium.CircleMarker(
          location=[df_select.iloc[i]['lat'], df_select.iloc[i]['long']],
          radius=np.sqrt(df_select.iloc[i]['Value'])*radius_scaling,
          popup='Location: ' + df_select.iloc[i]['Location'] + '\n Value: ' + str(df_select.iloc[i]['Value']),
          color='red',
          fill=True,
          fill_color='red',
          fill_opacity=0.25,
          weight=1 # stroke width in pixels
       ).add_to(my_map)

    return(my_map)
```

```python
#for each city, proportion of all kinds of pollution in 2020
my_map = plot_map(city_p2020,i_df=city_p2020, i_scale=1)
my_map
```
